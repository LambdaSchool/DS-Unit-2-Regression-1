```
import pandas as pd
from fbprophet import Prophet
import matplotlib.pyplot as plt
```

```
robo= pd.read_csv('https://raw.githubusercontent.com/alvinwalker314/DS-Unit-2-Regression-1/master/module1-making-forecasts/pageviews-20150701-20190523.csv')
```

```
robo.head()
robo.rename(columns= {'Date': 'ds', 'Robotics': 'y'}, inplace= True)
```

```
robo['Mean baseline']= robo['y'].mean() 
robo['Naive baseline'] = robo['y'].shift(1).ffill()
robo.plot(figsize= (15, 10))
```

```
model= Prophet()
```

```
model.fit(robo)
```

```
future= model.make_future_dataframe(periods= 365)
```

```
forecast= model.predict(future)
```

```
figure= model.plot(forecast)
plt.title('Robotics Page View Predictions', fontsize= 20)
plt.ylabel('total views', fontsize= 15)
plt.xlabel('Year')
```
