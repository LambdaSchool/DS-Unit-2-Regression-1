```python
gdpdf  = pd.read_csv('gdp.csv', header=None)
headers = gdpdf.T.iloc[1]
headers[0] = 'Year'
headers[1] = 'Quarter'
gdpdf = gdpdf.T[2:]
```

```python
gdpdf = gdpdf.rename(columns=headers)
gdpdf['Year'] = gdpdf['Year'].astype(int)
gdpdf.head()
```

```python
def get_sum_period(y1,y2, colname, df=gdpdf):
    df[colname] = df[colname].astype(float)
    sliced = df[(y1 <= df['Year']) & (df['Year'] <= y2)]
    return sliced[colname].sum()
    
```

```python
#current year and year prior
df['GDP Percent Increase'] = df['Year'].apply(lambda x: get_sum_period(x-1,x, 'Gross domestic product (GDP)', gdpdf))
```

```python
df.corr()[['Incumbent Party Vote Share']]
#GDP is a good predictor but not better than Average Recent Growth in Personal Incomes
```

```python
from sklearn.linear_model import LinearRegression
```

```python
parameters = df[['Average Recent Growth in Personal Incomes', 'US Military Fatalities per Million', 'GDP Percent Increase']]
y_true = df['Incumbent Party Vote Share']

model = LinearRegression().fit(parameters , y_true)
```

```python
pred = model.predict(parameters)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_true, pred)
```

```python
rpi_df = pd.read_csv('rpi.csv')
rpi_df['DATE'] = rpi_df['DATE'].apply(lambda x: x.split('-')[0]).astype(int)
rpi_df.head()
```

```python
def get_mean_pi(y1, y2, df=rpi_df):
    sliced = df[(y1 <= df['DATE']) & (df['DATE'] <= y2)]
    return sliced['DPCERL1A225NBEA'].mean()
```

```python
df['Real Persnal Income'] = df['Year'].apply(lambda x: get_mean_pi(x-1,x, rpi_df))

df.corr()[['Incumbent Party Vote Share']]
```
